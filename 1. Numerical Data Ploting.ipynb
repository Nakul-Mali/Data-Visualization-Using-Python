```python
import seaborn as sns
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# To set the style
sns.set(style='darkgrid') 
```

```python
# Load default dataset of seaborn libeary
tips = sns.load_dataset('tips') 
tips.head() 
```

```python
# relplot is a relational plot
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
# List of functions used on FacetGrid

dir(sns.FacetGrid) 
```

```python
# Total count of smoker and non smoker

tips['smoker'].value_counts() 
```

```python
# seprate this bill and tips with a smoker

sns.relplot(x='total_bill', y = 'tip', data = tips, hue = 'smoker')  
```
