```python
import seaborn as sns
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style='darkgrid') 
```

```python
# Load dataset
tips=sns.load_dataset('tips') 
tips.head() 
```

```python
sns.relplot(x='total_bill',y='tip',data=tips) 
```

```python
dir(sns.FacetGrid) 
```

```python
# Display bill based on the smoker

sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker') 
```

```python
# Count total no Smoker and Non Smoker

tips['smoker'].value_counts() 
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex') 
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='time') 
```

```python
# Visualize based on 4 parameters total_bill,tip,smoker and time

sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time') 
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='size') 
```

```python
# we can change the color by using palet 

sns.relplot(x='total_bill',y='tip',data=tips,hue='size',palette='ch: r=0.8,l=0.90')  
```

```python
# when total_bill and tip increses then size of circle increses

sns.relplot(x='total_bill',y='tip',data=tips,size='size') 
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='size',size='size') 
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,size='size',sizes=(15,200))  
```
